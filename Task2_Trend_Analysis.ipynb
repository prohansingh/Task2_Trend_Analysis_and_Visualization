```python
# Load dataset
df = pd.read_csv("C:/Users/dell/Documents/amazon_sales_data 2025.csv")
```

```python
print(df['Date'].head())
```

```python
# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
# Sort values by date
df = df.sort_values('Date')
```

```python
# Set date as index
df.set_index('Date', inplace=True)
```

```python
# Time-Series Line Plot

plt.figure(figsize=(12, 5))

plt.plot(df['Price'])

plt.title("Amount Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Amount")

plt.tight_layout()

plt.savefig("Task2 images/line_plot.png")

plt.close()
```

```python
import os

# Create folder automatically
os.makedirs("Task2 images", exist_ok=True)

# -----------------------------------
# Moving Average
# -----------------------------------

df['Moving_Average'] = df['Total Sales'].rolling(window=7).mean()

plt.figure(figsize=(12, 5))

plt.plot(df['Total Sales'], label='Original Sales')

plt.plot(df['Moving_Average'], label='7-Day Moving Average')

plt.title("Sales Trend with Moving Average")

plt.xlabel("Date")

plt.ylabel("Total Sales")

plt.legend()

plt.tight_layout()

plt.savefig("Task2 images/moving_average.png")

plt.close()
```

```python
monthly_sales = df.groupby(df.index.month)['Total Sales'].sum()

plt.figure(figsize=(10,5))

monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.tight_layout()

plt.savefig("Task2 images/monthly_sales_trend.png")

plt.close()
```
